```python
from bs4 import BeautifulSoup
import requests

base_url = 'https://lifehacker.ru/topics/technology/all'
parsed_urls = []  # Инициализируем список перед использованием

for page_num in range(1, 11):
    url = f'{base_url}/page{page_num}'  # Формируем URL страницы
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    raw_items = soup.find_all('a', class_='tm-article-comments-counter-link__link')

    # Добавляем обработанные URL в список
    parsed_urls += [item.get('href').replace('/comments', '') for item in raw_items]

# Выводим результат
print(f"Всего собрано URL: {len(parsed_urls)}")
print("Примеры URL:", parsed_urls[:5])  # Показываем первые 5 URL
```

```python
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm  # Импортируем tqdm для прогресс-бара

result = []

for url in tqdm(parsed_urls):  # Теперь tqdm определен
    try:
        article = {}
        article_url = f'https://habr.com{url}'  # Исправлено: добавляем полный домен
        headers = {'User-Agent': 'Mozilla/5.0'}

        response = requests.get(article_url, headers=headers, timeout=10)
        response.raise_for_status()  # Проверка на ошибки HTTP

        soup = BeautifulSoup(response.text, 'lxml')

        # Безопасное извлечение данных с проверками
        title_tag = soup.find('h1', class_='tm-title_h1')
        article['title'] = title_tag.span.text if title_tag and title_tag.span else 'Нет заголовка'

        article_body = soup.find('div', class_='tm-article-body')
        article['text'] = article_body.text if article_body else 'Нет текста'

        votes_counter = soup.find('span', class_='tm-votes-lever__score-counter')
        article['votes_count'] = votes_counter.text.strip('+') if votes_counter else '0'

        result.append(article)

    except Exception as e:
        print(f"Ошибка при обработке {url}: {str(e)}")
        continue  # Пропускаем проблемную статью

# Вывод результатов
print(f"\nОбработано статей: {len(result)}")
print("Пример первой статьи:", result[0] if result else "Нет данных")
```

```python
import requests
from bs4 import BeautifulSoup

# URL статьи для парсинга
url = 'https://lifehacker.ru/kak-ochistit-ekran-televizora-bez-razvodov/'

# Заголовки для имитации браузера
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    # Получаем содержимое страницы
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Проверяем на ошибки HTTP

    soup = BeautifulSoup(response.text, 'lxml')

    # Парсим заголовок (новый селектор)
    title = soup.find('h1', class_='_1Zk2X').get_text(strip=True) if soup.find('h1', class_='_1Zk2X') else 'Не удалось получить заголовок'

    # Парсим основной текст статьи (новый селектор)
    article_body = soup.find('div', class_='_3O2VK')
    if article_body:
        # Удаляем ненужные элементы (реклама, рекомендации)
        for element in article_body.find_all(['div', 'aside', 'figure', 'script']):
            element.decompose()
        text = article_body.get_text('\n', strip=True)
    else:
        text = 'Не удалось получить текст статьи'

    # Парсим дату публикации (новый селектор)
    date = soup.find('time', class_='_2kXAl').get_text(strip=True) if soup.find('time', class_='_2kXAl') else 'Не удалось получить дату'

    # Выводим результаты
    print(f"Заголовок: {title}")
    print(f"\nДата публикации: {date}")
    print("\nТекст статьи:")
    print(text)

except Exception as e:
    print(f"Произошла ошибка: {e}")
```
